## Final

---------------- Méthodologie --------------

Les traitement appliqués à ce type de structure :

- Parcourir la liste et extraire chaque ligne de la colonne modification.

 - Trier par les listes de modifications par dates si disponnible ou par 'id' pour la dernière syntaxe utilisé dans les données. Si aucune indication disponnible, appliquer les modifications dans l'ordre. 

 - Dans les cas, nouvelle syntaxe, Ignorer les modification['id': str] comme clé de tri.

 ##### Si titulaires existe : __Pas encore implémenté !!!__

 - Parcourir la liste de dictionnaire (Attention 2 format différents, voir données problématiques)

 - Aplatir les dicts imbriqués avec préfixe (titulaire_)

 - Appliquer les modifications et SI plusieurs tutilaires ne pas les remplacer mais tous les ajouter.

```python
""" !!!! FINAL !!!!

Version fonctionnel pour la syntaxe linéaire de la colonnes modifications.

Mais inutile sur les syntaxe avec des dictionnaires et des listes imbriquées
(allé voir : "exemples données problématiques" en fin de note book)
Piste à exploiter dans les commentaires de ce code 
"""

modifications_invalides = 0
exemples = []

def appliquer_modifications(row):
    global modifications_invalides  # Important pour modifier la variable globale

    try:
        modifications_raw = row["modifications"]
        modifications_list = json.loads(modifications_raw)
    except (json.JSONDecodeError, TypeError):
        modifications_invalides += 1  # Incrémenter le compteur d'erreurs
        exemples.append(row.get("modifications"))
        return row  # on ignore si la valeur est vide ou mal formée

    # Si ce n'est pas une liste (ex: un dict direct ou None), on ignore
    if not isinstance(modifications_list, list):
        #modifications_invalides += 1  # 61 lignes invalides !!!  donc potentiellement non traitées mais je ne sais pas vraiment ce qui bloque 
        #exemples.append(row.get("modifications"))
        return row

    # Extraire proprement les dicts de modification (cas [{"modification": {...}}, ...])
    #  En l'adaptant, potentiellement réutilisable pour l'extraction des listes de dicts des titulaires
    modifs = [
        mod["modification"] if isinstance(mod, dict) and "modification" in mod else mod
        for mod in modifications_list
    ]

    # Tri personnalisé sans convertir les id entiers en chaînes
    def tri_modification(mod):
        return (
            mod.get("datePublicationDonneesModification")
            or mod.get("dateNotificationModification")
            or mod.get("dateSignatureModification")
            or mod.get("updated_at")
            or (mod.get("id") if isinstance(mod.get("id"), str) else "")
        )

    modifs_sorted = sorted(modifs, key=tri_modification)

    for modif in modifs_sorted:
        for key, value in modif.items():

            # Gestion spéciale de 'id'
            if key == "id":
                if isinstance(value, int):
                    continue  # On ignore les id techniques
                elif isinstance(value, str):
                    row["id"] = value  # Appliquer le vrai identifiant
                    continue

            # Normalisation du nom de la clé
            base_key = (
                key.replace("Modification", "") if key.endswith("Modification") else key
            )

            if base_key in row:
                try:
                    if isinstance(row[base_key], (float, int)):
                        row[base_key] = float(value)
                    else:
                        row[base_key] = value
                except:
                    row[base_key] = value

    return row


# Application sur le DataFrame
df_test = df_MP_brut.apply(appliquer_modifications, axis=1)
print(f"Nombre de lignes avec modifications invalides : {modifications_invalides}")

for i in exemples:
    print(i)
```

```python
""" Modifications simple, chronologique OUI
imbrication non supporté, titulaires : null non supporté
Problème correspondance de id et id non résolus 
mais tente de gérer les imbrications titulaires"""

from datetime import datetime

# À stocker globalement si tu veux traquer les erreurs
modifications_invalides = 0
exemples_invalides = []

# Liste des clés qu'on peut utiliser pour trier les modifs
DATE_KEYS = ["datePublicationDonnees", "dateNotification", "dateSignature", "updated_at"]

def extraire_date(modif):
    for key in DATE_KEYS:
        val = modif.get(key)
        if val:
            try:
                return datetime.fromisoformat(val)
            except Exception:
                pass
    return datetime.min  # pour forcer tri en dernier si date absente

def aplatir_dictionnaire(d, parent_key=''):
    """Aplatit un dictionnaire imbriqué avec noms de colonnes adaptés."""
    items = {}
    for k, v in d.items():
        new_key = f"{parent_key}_{k}" if parent_key else k
        if isinstance(v, dict):
            items.update(aplatir_dictionnaire(v, new_key))
        else:
            items[new_key] = v
    return items

def appliquer_modifications_complexes(row):
    global modifications_invalides, exemples_invalides
    raw = row.get("modifications")

    if not raw:
        return row

    try:
        # 1. Parsing JSON (ou plusieurs JSON imbriqués)
        if isinstance(raw, str):
            modifs = json.loads(raw)
        else:
            modifs = raw

        # 2. Normalisation vers liste
        if isinstance(modifs, dict):
            modifs = [modifs]
        elif isinstance(modifs, list):
            parsed = []
            for item in modifs:
                if isinstance(item, dict):
                    parsed.append(item)
                elif isinstance(item, str):
                    try:
                        parsed.append(json.loads(item))
                    except Exception:
                        pass
            modifs = parsed
        else:
            raise ValueError("Type non pris en charge")

    except Exception:
        modifications_invalides += 1
        exemples_invalides.append(raw)
        return row

    # 3. Extraction des vraies modifications
    modifications_flat = []

    for modif in modifs:
        # Cas 3 : dictionnaire avec "modification" dedans
        if isinstance(modif, dict) and "modification" in modif:
            m = modif["modification"]
            if isinstance(m, dict):
                modifications_flat.append(m)
        else:
            modifications_flat.append(modif)

    # 4. Tri des modifs par date (plus ancienne → plus récente)
    modifications_flat.sort(key=extraire_date)

    # 5. Application des modifications
    for modif in modifications_flat:
        for key, val in modif.items():
            if val in (None, "", []):
                continue

            # Traitement des titulaires imbriqués
            if key == "titulaires":
                # Peut être un dict ou une liste
                t_list = modif["titulaires"]
                if isinstance(t_list, dict):
                    t_list = [t_list]
                for t in t_list:
                    if "titulaire" in t:
                        flat = aplatir_dictionnaire(t["titulaire"], parent_key="titulaire")
                        for col, v in flat.items():
                            if col in row:
                                row[col] = v
                continue

            # Ignorer le champ "id" quand il ne correspond pas à la colonne "id"
            if key == "id" and "modification" in modif:
                continue

            # Clé directe (clé == nom de colonne)
            if key in row:
                row[key] = val
                continue

            # Clé de type "nomColonneModification"
            if key.endswith("Modification"):
                base_key = key[:-12]
                if base_key in row:
                    row[base_key] = val

    return row


df_test = df_MP_manip.apply(appliquer_modifications_complexes, axis=1)

print("Modifications invalides :", modifications_invalides)
for ex in exemples_invalides[:3]:
    print(ex)


```
