La REGEX dure 35 min avec mon PC, à voir si on veut la faire tourner dans la pipeline

```python
import re

# Helper function to split text into words, filter words with more than 3 letters, and remove the last letter
def split_filter_and_trim_words(text):
    if pd.notna(text):
        return [unidecode(word[:-1].lower()) for word in re.findall(r'\w+', text) if len(word) > 3]
    return []

# Apply the helper function to split, filter, and trim words in both columns
liste_mandat['filtered_qualite'] = liste_mandat['matching_qualite'].apply(split_filter_and_trim_words)
df_mandat['filtered_description'] = df_mandat['matching_description'].apply(split_filter_and_trim_words)

# Create a new column in df_mandat to store matched 'qualite' values
df_mandat['matched_qualite'] = None

# Iterate through each row in liste_mandat
# and check for matches in df_mandat
for _, liste_mandat_row in liste_mandat.iterrows():
    filtered_qualite = liste_mandat_row['filtered_qualite']
    original_qualite = liste_mandat_row['matching_qualite']  # Keep the original qualite for reporting
    elu_liste_mandat = liste_mandat_row['élu']
    if filtered_qualite:  # Ensure there are words to match
        # Iterate through each row in df_mandat
        for mandat_index, mandat_row in df_mandat.iterrows():
            filtered_description = mandat_row['filtered_description']
            elu_mandat = mandat_row['élu']
            # Check if 'élu' keys match
            if elu_liste_mandat == elu_mandat and filtered_description:
                # Count matches for words in filtered_qualite
                match_count = sum(
                    any(re.search(rf'\b{word}\b', desc_word, re.IGNORECASE) for desc_word in filtered_description)
                    for word in filtered_qualite
                )
                # Check if matches are found for all or all but one word
                if match_count >= len(filtered_qualite) - 1:
                    # Report the original qualite in df_mandat
                    df_mandat.at[mandat_index, 'matched_qualite'] = original_qualite

# Display the updated df_mandat
df_mandat
```

L'élu a l'obligation de déclarer ses revenus et patrimoines en début de mandat, puis à tout changement de sa situation (ie. chaque année), source : https://www.hatvp.fr/consulter-les-declarations/ .

Le calcul du barème se base sur la liste de publication des déclarations, autrement dit leurs en-têtes, pas leur contenu.

L'élu peut être dispensé de toute déclaration, listé dans la colonne statut_publication du dataframe liste.
Si l'élu n'a pas déclaré alors qu'il est obligé, statut_publication = 'déclaration non déposée. On lui attribue le Score E

```python
# Ajouter une nouvelle colonne avec une logique conditionnelle
listed_score['score_E'] = listed_score['statut_publication'].str.lower().apply(
    lambda x: 1 if x == 'déclaration non déposée' else 0)

listed_score
```

On rassemble les types de déclarations : même identification (DI ou DSP) des déclarations initales et modificatives, par année. 

```python
listed_score['type_decla'] = np.where(listed_score['type_document'].str.startswith('di', na=False), 'di',
    np.where(listed_score['type_document'].str.startswith('dsp', na=False), 'dsp', None)
)

listed_score['total_type_document'] = listed_score.groupby(['année','key'])['type_decla'].transform('nunique')
listed_score
```

L'élu doit effectuer une déclaration de type DI par année selon son obligation de publication. Revoir https://www.hatvp.fr/wordpress/wp-content/uploads/2024/07/Tableau-recapitulatif-des-declarants-20240726.pdf 

Si l'élu a effectivement rempli son obligation, il a un score C. Déclaration(s) effectuée(s) = Déclaration(s) obligatoire(s).

Si l'élu a rempli son obligation mais a signalé tout changement en anticipation de l'année suivante, on lui attribue le Score B. Déclaration(s) effectuée(s) > Déclaration(s) obligatoire(s)


```python
# Group by 'élu' and 'année' and calculate the count of 'type_document'
type_document_count = listed_score.groupby(['key', 'année'])['type_document'].count()

# Compare the count with 'total_type_document' for each 'élu' and 'année'
def score_C(row):
    count_C = type_document_count.get((row['key'], row['année']), 0)
    return 1 if count_C == row['total_type_document'] else 0

def score_B(row):
    count_B = type_document_count.get((row['key'], row['année']), 0)
    return 1 if count_B > row['total_type_document'] else 0

# Apply the function to calculate 'score_C'
listed_score['score_C'] = listed_score.apply(score_C, axis=1)
# Apply the function to calculate 'score_B'
listed_score['score_B'] = listed_score.apply(score_B, axis=1)
listed_score
```

Score A = fonction du join df_mandat avec liste via la regex donc seulement 10% des élus sont notés

On calcule la différence de 60 jours entre date de dépôt obligatoire et date de début de mandat, en soustrayant l'année de déclaration (1 déclaration par année)
Si c'est inférieur à 60 jours, on augmente le barème de 1 : 
- Si l'élu avait C -> sa note passe à B
- Si l'élu avait B -> sa note passe à A

Il faut donc cumuler les bonnes pratiques pour avoir A

```python
from datetime import timedelta

listed_score['score_A'] = listed_score.apply(
    lambda row: 1 if pd.notna(row['Date_début_Mandat']) and pd.notna(row['date_depot']) and 
    abs((row['Date_début_Mandat'].replace(year=row['année']) - row['date_depot']).days) <= 60 else 0,
    axis=1
)
listed_score
```

```python
# Replace conditions with explicit if statements
def calculate_score_global(row):
    if row['score_E'] == 1:
        return 'E'
    elif row['score_C'] == 1 and row['score_B'] == 0 and row['score_A'] == 0:
        return 'C'
    elif (row['score_B'] == 1 and row['score_A'] == 0) or (row['score_C'] == 1 and row['score_B'] == 0 and row['score_A'] == 1):
        return 'B'
    elif (row['score_B'] == 1 and row['score_A'] == 1):
        return 'A'
    else:
        return 'C'

# Apply the function to calculate 'score_global'
listed_score['score_global'] = listed_score.apply(calculate_score_global, axis=1)

# update score_global si dispensé de toute déclaration
listed_score['score_global'] = np.where(listed_score['statut_publication'].str.lower() == 'dispense', 'dispense', listed_score['score_global'])
listed_score
```

Group by pour avoir la tendance du score_global par année pour chaque élu

```python
listed_score_élu = listed_score.groupby(['nom', 'prenom', 'année', 'élu'], as_index=False).agg({
    'score_global': lambda x: x.mode()[0] if not x.mode().empty else None,
    'Date_début_Mandat': 'min',
})
listed_score_élu
```
