```python
liste['élu'] = liste['nom'].str.lower() + "_" + liste['prenom'].str.lower() + "_" + liste['type_mandat'].str.lower()
df_mandat['élu'] = df_mandat['Nom'].str.lower() + "_" + df_mandat['Prénom'].str.lower() + "_" + df_mandat['Organe_Intitulé_Mandat'].str.lower()


listed = pd.merge(liste, df_mandat[["élu",'Date_début_Mandat']], on= 'élu', how="left")
listed
```

```python
listed['année'] = pd.to_datetime(listed['date_depot'], errors='coerce').dt.year
listed['année'] = pd.to_datetime(listed['année'], format='%Y', errors='coerce')
```

```python
# Ajouter une nouvelle colonne avec une logique conditionnelle
listed['score_E'] = listed['statut_publication'].str.lower().apply(
    lambda x: 1 if x == 'déclaration non déposée' else 0)

listed
```

```python
listed['type_decla'] = np.where(
    listed['type_document'].str.startswith('di', na=False), 'di',
    np.where(listed['type_document'].str.startswith('dsp', na=False), 'dsp', None)
)

listed['total_type_document'] = listed.groupby(['année', 'prenom', 'nom'])['type_decla'].transform('nunique')
listed
```

```python
# Group by 'élu' and 'année' and calculate the count of 'type_document'
type_document_count = listed.groupby(['élu', 'année'])['type_document'].count()

# Compare the count with 'total_type_document' for each 'élu' and 'année'
def score_C(row):
    count_C = type_document_count.get((row['élu'], row['année']), 0)
    return 1 if count_C == row['total_type_document'] else 0

def score_B(row):
    count_B = type_document_count.get((row['élu'], row['année']), 0)
    return 1 if count_B > row['total_type_document'] else 0

# Apply the function to calculate 'score_C'
listed['score_C'] = listed.apply(score_C, axis=1)
# Apply the function to calculate 'score_B'
listed['score_B'] = listed.apply(score_B, axis=1)
listed
```

```python

listed['note_intention'] = listed['total_note_publication']/listed['total_type_document']
listed['note_intention'] = np.where(listed['statut_publication'].str.lower() == 'dispense', 1, listed['note_intention'])
listed
```

```python
conditions = [
    listed['score_E'] == 1,
    listed['score_D'] == 1,
    listed['score_C'] == 1,
    listed['score_B'] == 1,
    listed['score_A'] == 1
]

# Define corresponding values
choices = ['E', 'D', 'C', 'B', 'A']

# Apply conditions to create the 'score_total' column
listed['score_total'] = np.select(conditions, choices, default='C')
```
